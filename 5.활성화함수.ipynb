# 개요

- activation function

  - **정확도를 높이기 위해서** 다양한 방법들이 시도된다. 그런 방법들 중에 하나 활성화 함수를 활용하는것이다

# sigmod

- **초기에 딥러닝시 가장 많이 사용**되었다
- 입력 : 실수값
- 출력 : 0 ~ 1사이 값으로 출력
  - 어떤 입력이 와도 음수값은 존재하지 않는다
  - 단, x(입력) 커지거나 (1.0 수렴), 아주 작아지면(0.0 수렴) 
    - 기울기가 0에 수렴한다
      - 가중치가 미반영된다
        - 학습이 않되고 있다 
          - 신경이 죽았다(뉴런이 죽었다)
            - **죽은 뉴런 문제!!**

# Relu

- 최근 가장 많이 사용한다
- 하위 활성화 함수들도 많이 생성
- 장점
  - 최적화 도구(SGD, Adam, Momentom,..)가 빠르게 최적의 가중치를 찾도록 도와준다
  - 연산 비용이 저렴하다
    - x가 양수면 그대로,  x가 음수면 무조건 0
- 단점
  - 오차역전파(y->x) 진행시, 기울기가 너무 커져버리면, 무응답상태가 올수도 있다
    - 학습(X) -> 죽은 뉴런 문제 발생!!
