# 개요

- 생성적 적대 신경망
- **생성 모델**의 대표 신경망
  - 창조!!, 이를 기반으로 응용
  - 가장 혁신적인 딥러닝 분야
- Generatibe Adversarial Network


- 기본적으로 동시에 2개의 모델을 훈련
  - 생성자 (Generative)
    - 가짜 데이터를 생성하도록 훈련
    - 판별자를 잘 속이는게 목적
      - 거의 진짜와 유사하도록 가짜 데이터를 만드는것 목적
      - 뭉크 작품을 훼이크 하고 싶다면 -> 뭉크 데이터셋 에 대한 훈련이 필요하다
  - 판별자 (Discriminant)
    - 실제 데이터와 가짜 데이터를 잘 구분하도록 훈련
    - 목표: 진짜를 잘 구분하는것

- 적대적 (Adversarial)
  - GAN의 뼈대를 이루는 두개의 모델이 게임같은 경쟁구도를 표현하는것이다
  - 생성자는 실제 데이터와 구분이 않될 정도로 유사한 샘플을 만드는것이다
    - 이렇게 훈련된 모델로 만든 데이터를 이용
    - 데이터가 부족한 다른 분야에 데이터를 생성
      - 데이터 부족분 해결 가능
  - 판별자
    - 실제가 가짜를 잘 구분이 목표
    - 더이상 구분할수 없을때 종료

# 생성자

- 입력
  - 랜덤한 잡음(숫자)로 구성된 백터 z
- 출력
  - 최대한 진짜로 보이는 가짜 샘플
- 목표
  - 훈련 데이터셋에 있는 진짜 샘플과 구분이 불가능한 가짜 샘플을 생성

  - 입력 샘플이 진짜일 예측 확률
- 목표
  - 진짜와 가짜를 잘 구분하는 것

# 종류

- AutoEncoder
  - 자동으로 데이터를 인코딩 하게 도와주는 모델
  - 인코더 <-> 디코더
  - 2개가 경쟁하면서 발전하는 모델이며, GAN의 기본 아이디어를 제공한 모델

- GAN
  - 아래의 모든 모델들의 원류(조)

- DCGAN
  - 모델의 성능을 높이자 
  - 판별자 모델을 합성곱 신경망으로 대체해서 구현
  - Deep Convolution GAN
  - 최신 GAN 모델의 기초 성립

- WGAN

- ProGAN
  - 풀 HD 화질로 실제 사진과 같은 이미지를 생성하는 기법
  - 전이학습으로 사용가능

- SGAN
  - 준지도 학습 분야 기대되는 모델
    - 레이블 데이터가 동비율로 존재할경우(절반은 정답이 있고, 나머지는 없다)

- CGAN

- CycleGAN
  - 이미지(비디오)대 이미지(비디오) 변환
  - 스타일 트렌스퍼
  - 화풍을 가져와서 새로운 사진에 대입, 마치 그 화가가 이 이미지를 그린것처럼 훼이크 데이터를 생성하는 방식
